```python
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
factory = StemmerFactory()
stemmer = factory.create_stemmer()
stemmed_tweet=[]
for i in df['cleaned_content']:
    stemmed_tweet.append(stemmer.stem(i))
df['cleaned_content']=stemmed_tweet
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
count=Counter()
for i in df1['cleaned_content']:
    for j in i:
        count[j]+=1
Word=[]
Count=[]
for i in count.most_common(15):
    Word.append(i[0])
    Count.append(i[1])
plt.figure(figsize=(20,4))
sns_plot = sns.barplot(Word,Count)
```
